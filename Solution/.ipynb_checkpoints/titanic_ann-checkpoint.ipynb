## 2.特征选择

```python
all_features.drop(['Name'],axis=1,inplace=True) # 删除一些不需要的特征
all_features
```

```python
#1 标准化
numeric_features_index = all_features.dtypes[all_features.dtypes != 'object'].index
all_features[numeric_features_index] = all_features[numeric_features_index].apply(
    lambda x: (x - x.mean()) / (x.std()))
all_features
```

```python
all_features[numeric_features_index].isnull().sum()
```

```python
all_features = pd.get_dummies(all_features) # 一个列里面有5个不同的字符串，则创建5个feature
all_features
```

```python
#2 “数值型”特征填充缺失值
all_features = all_features.dropna(subset=['Fare'],axis=0)#删除含有缺失值的特定的行
all_features[numeric_features_index].isnull().sum()
```

```python
# 用0填充NA值
all_features_zero=all_features.copy()
all_features_zero[numeric_features_index] = all_features[numeric_features_index].fillna(0) 
all_features_zero[numeric_features_index].isnull().sum()
```

```python
all_features_mean=all_features.copy()
all_features_mean[numeric_features_index] = all_features[numeric_features_index].fillna(all_features[numeric_features_index].median()) 
all_features_mean[numeric_features_index].isnull().sum()
```

从data_frame转换为tensor

```python
n_train = train.shape[0]#训练集的个数
train_features = torch.tensor(all_features_mean[:n_train].values, dtype=torch.float32)
test_features = torch.tensor(all_features_mean[n_train:].values, dtype=torch.float32)
train_labels = torch.tensor(train.Survived.values.reshape(-1, 1), dtype=torch.int64)#train.Survived是否获救的label
train_labels = torch.zeros(891,2).scatter_(1,train_labels,1)
train_features.shape
```

## 3.模型训练

### 3.1设置超参数 

```python
lr = 0.01
gamma = 0
epochs = 10
bs = 128
weight_decay=0.1
k=5
```

### 3.2分割小批量

```python
from torch.utils import data
```

```python
dataset = data.TensorDataset(*(train_features, train_labels))
batchdata = DataLoader(dataset,batch_size=bs,shuffle = True)
len(batchdata)
```

```python
input_ = train_features.data[0].numel() #特征的数目，一般是第一维之外的所有维度相乘的数
output_ = len(train_labels.unique()) #分类的数目
output_
```

### 3.3定义神经网络框架

```python
class Model(nn.Module):
    def __init__(self,in_features,out_features):
        super().__init__()
        self.linear1 = nn.Linear(in_features,128,bias=True)
        self.linear2 = nn.Linear(128,128,bias=True)
        self.output = nn.Linear(128,output_,bias=True)
        
    def forward(self,x):
        sigma1 = torch.relu(self.linear1(x))
        z2 = self.linear2(sigma1)
        sigma2 = torch.relu(z2)
        z3 = self.output(sigma2)
        sigma3 = torch.sigmoid(z3)
        return sigma3
```

### 3.4定义训练函数

```python
def trans_to_onehot(X):
    result=[]
    for i in range(len(X)):
        max=0
        max_index=0
        result2=[]
        for j in range(len(X[i])):
            if X[i][j]>max:
                max=X[i][j]
                max_index=j
        for k in range(len(X[i])):
            if k !=max_index:
                result2.append(0)
            else:
                result2.append(1)

        result.append(result2)
    result= torch.tensor(result)
    return result
```

```python
def fit(net,batchdata,lr=0.01,epochs=5,weight_decay=0.1,gamma=0):
    criterion = nn.BCELoss() #定义损失函数
    correct = 0
    samples = 0
    
    opt = optim.SGD(net.parameters(), lr=lr,momentum=gamma)

    for epoch in range(epochs):
        for batch_idx, (x,y) in enumerate(batchdata):

            sigma = net.forward(x)
            loss = criterion(sigma,y)
            loss.backward()
            opt.step()
            opt.zero_grad()
            
            yhat = trans_to_onehot(sigma)

            correct += torch.sum(yhat == y)/2
            samples += x.shape[0]
            

        if (batch_idx+1) % 125 == 0 or batch_idx == len(batchdata)-1:
            print('Epoch{}:[{}/{}({:.0f}%)]\tLoss:{:.6f}\t Accuracy:{:.3f}'.format(
                                                                                    epoch+1
                                                                                    ,samples
                                                                                    ,len(batchdata.dataset)*epochs
                                                                                    ,100*samples/(len(batchdata.dataset)*epochs)
                                                                                    ,loss.data.item()
                                                                                    ,float(correct*100)/samples))
            
        
                
        
```

```python
net = Model(in_features=input_, out_features=output_)
```

```python
fit(net,batchdata,lr,epochs,gamma)
```
